```python
# Libraries
## Mathematics
import numpy as np
from time import time

## Visualization
import matplotlib.pyplot as plt
```

```python
# Load functions .py files
from Checking_simulated_boundary_functions_laws import optimal_stopping_montecarlo_laws
```

# Brownian bridge

```python
# Initialize variables
r = 0 #Pinning point
mesh = 1000 # Points in temporal grid
N = mesh
L = 10000 # Points in the spatial grid
M = 4000 # Monte Carlo iterations
mu = "discrete" # Type of mixture
weights = np.array([1]) # Weights in the mixture
parameters = np.array([[r],[1]]) # Parameters of each distribution in the mixture
a = -1 # Lower bound of the spatial grid
b = 1 # Upper bound of the spatial grid
alpha = 2 # # Scaling parameter


# Boundary obtention
start = time()
np.random.seed(123)
value_function, X_vals, parallel_time = optimal_stopping_montecarlo_laws(mu = mu, 
                                                     weights = weights, 
                                                     parameters = parameters, 
                                                     N = mesh, 
                                                     a = a, 
                                                     b = b, 
                                                     L = L, 
                                                     M = M,
                                                     alpha = alpha)
end = time()
total_time = end - start
p = parallel_time/total_time
```

## Amdahl's law

```python
N = np.arange(1, 101)

# Function
f_Amdahl = 1/((1-p) + p/N)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(N, f_Amdahl, '-', color='blue')
plt.title("Amdahl's law for the Brownian Bridge case")
plt.xlabel("Number of cores")
plt.ylabel("Maximum Speedup")
plt.grid(True)
plt.xticks(np.arange(20, 101, 20))
plt.tight_layout()
plt.savefig("Amdahl_BB.png", dpi=300)
plt.show()
```

## Gustafson's law

```python
N = np.arange(1, 101)

# Function
f_Gustafson = (1 - p) + p * N

# Plot
plt.figure(figsize=(8, 5))
plt.plot(N, f_Gustafson, '-', color='blue')
plt.title("Gustafson's law for the Brownian Bridge case")
plt.xlabel("Number of cores")
plt.ylabel("Maximum Speedup")
plt.grid(True)
plt.xticks(np.arange(20, 101, 20))
plt.tight_layout()
plt.savefig("Gustafson_BB.png", dpi=300)
plt.show()
```

```python
# Libraries
## Mathematics
import numpy as np
from scipy.stats import norm
from scipy.interpolate import interp1d

## Visualization
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
# Load functions .py files
from Theoretical_boundaries_visualization_functions import optimal_stopping_Brownian_bridge, optimal_stopping_Normal
from Visualisation import OSB_visualisation
from Checking_simulated_boundary_functions_3 import optimal_stopping_montecarlo_3
```

# Exact optimal stopping boundary for a degenerated distribution

```python
# Pinning point
r = 0

# Temporal mesh
mesh = 1000
t_mesh = np.linspace(0, 1, mesh)

# Boundary
exact_boundary_BB = optimal_stopping_Brownian_bridge(t_mesh, r)
L = 10000
np.random.seed(123)
value_function_BB, X_vals_BB = optimal_stopping_montecarlo_3(mu = "discrete", weights = np.array([1]), parameters = np.array([[r],[1]]), N = mesh, a = -1, b = 1, L = L, M = 4000, alpha = 2)

# Plotting
OSB_visualisation(N = mesh, value_function = value_function_BB, X_vals = X_vals_BB, title = f"Optimal stopping boundary for degenerated distribution in r={r}", name = "BB_MonteCarlo.png", exact_boundary = exact_boundary_BB)
```

# Theoretical optimal stopping boundary for a Normal distribution with standard deviation lower than 1

## Standard deviation of 1/2

```python
# Normal distribution setup
mean = 0
sigma = 1/2

# Temporal mesh
mesh = 1000
t_mesh = np.linspace(0, 1, mesh)

# Boundary
exact_boundary_Normal = optimal_stopping_Normal(mesh=t_mesh, m=mean, gamma=sigma, tol=1e-6)
L = 10000
np.random.seed(123)
value_function_Normal, X_vals_Normal = optimal_stopping_montecarlo_3(mu="continuous", weights=np.array([1]), parameters=np.array([[mean], [sigma]]), N=mesh, a=-1.3, b=1.3, L=L, M=4000, alpha = 1.5)


# Plotting
OSB_visualisation(N = mesh, value_function = value_function_Normal, X_vals = X_vals_Normal, title = fr"Optimal stopping boundary for N({mean}, {sigma})", name = "Normal_MonteCarlo.png", exact_boundary = exact_boundary_Normal)
```
